```
import pandas as pd
import requests
import os
import json
import csv


def get_webpage(webpage_url):
    # Returns html for the page provided. Input: website url. Output: raw html
    page = requests.get(webpage_url)
    return(page)
```

```
#go out and get the Florida Covid data

webpage_url = "https://services1.arcgis.com/CY1LXxl9zlJeBuRZ/arcgis/rest/services/Florida_COVID19_Case_Line_Data_NEW/FeatureServer/0/query?where=1%3D1&outFields=EDvisit,Hospitalized,Died,Case_,Contact,Case1,EventDate,ChartDate,ObjectId&outSR=4326&f=json"  
        
web_page = get_webpage(webpage_url)
```
